```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the dataset using the raw URL
df = pd.read_csv('https://raw.githubusercontent.com/PPARIHAR0027/EY_PARRRK_Group2/main/LaptopPrice.csv')
df.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 2. Prepare the features (X) and target (y)
X = df[['Ram', 'PrimaryStorage', 'CPU_freq', 'ScreenW']]  # Features
y = df['Price_euros']  # Target variable

# 3. Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 4. Scale the features (Standardization helps neural networks converge faster)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 5. Build the neural network model
model = Sequential()

# Input layer with 4 neurons (one for each feature) and a hidden layer with 64 neurons
model.add(Dense(64, input_dim=4, activation='relu'))

# Add more hidden layers if needed
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))

# Output layer with 1 neuron (since we're predicting price)
model.add(Dense(1))

# 6. Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# 7. Train the model
model.fit(X_train, y_train, epochs=200, batch_size=10, validation_split=0.2)

# 8. Evaluate the model on the test set
loss = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss}")

# 9. Make predictions
predictions = model.predict(X_test)

# 8. Calculate performance metrics
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R²): {r2}")

# Example of printing a few predictions alongside actual values
for i in range(10):
    print(f"Predicted Price: {predictions[i][0]}, Actual Price: {y_test.iloc[i]}")
```
